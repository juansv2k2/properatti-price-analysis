```python
def plots(y_test, y_pred, title):
    #definimos los residuales para los gráficos
    residuales = y_test - y_pred

    #graficamos el scatter de predicciones vs actuales
    plt.scatter(y_test, y_pred)
    plt.plot(y_test, y_test, color='red', linestyle='-')  #diagonal de pendiente 1
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs. Predicted Values '+ title)
    plt.show()

    #graficamos el scatter de residuales
    plt.figure(figsize=(10, 6))
    plt.scatter(np.arange(len(residuales)), residuales)
    plt.axhline(y=0, color='r')
    plt.title('Scatter de Residuales '+ title)
    plt.xlabel('Índice de muestra')
    plt.ylabel('Residuales')
    plt.show()

    #graficamos el histograma de residuales y superponemos el gráfico de la función normal
    plt.figure(figsize=(10, 6))
    plt.hist(residuales, bins=30, edgecolor='black', alpha=0.7, label='Histograma de Residuales')
    mean, std_dev = np.mean(residuales), np.std(residuales)
    x = np.linspace(min(residuales), max(residuales), 100)
    y = norm.pdf(x, mean, std_dev)
    y = y * len(residuales) * np.diff(np.histogram(residuales, bins=30)[1])[0]
    plt.plot(x, y, 'r--', label='Distribución Normal')
    plt.axvline(x=0, color='r', linestyle='--')
    plt.title('Histograma de Residuales '+ title)
    plt.xlabel('Residuales')
    plt.ylabel('Frecuencia')
    plt.show()

    stats.probplot(residuales, dist="norm", plot=plt)
    plt.title('QQ Plot ' + title)
    plt.show()
    return
```

```python
# Results: Comparison of Model Residuals and Predictions

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Helper function to collect residuals and predictions
def get_residuals_and_preds(model_func, X, y, title):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8972)
    model = model_func()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    residuales = y_test - y_pred
    return y_test, y_pred, residuales, title

# Define features for each model
features_simple = ["surface_total_clean_normalized", "rooms_clean_normalized", "precios_nuevos_normalized"]
features_full = features_simple + ["estacionamiento", "sum", "amenities", "pileta", "parrilla", "gimnasio", "lavadero", "solarium", "quincho", "baulera", "terraza"]

# Prepare data for Palermo
X_simple = properatti_palermo[features_simple]
X_full = properatti_palermo[features_full]
y = properatti_palermo.price_usd_per_m2

# Collect results for Linear, Lasso, Ridge (simple and full)
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV

results = []

# Linear Regression (simple)
results.append(get_residuals_and_preds(lambda: LinearRegression(), X_simple, y, "Linear (Simple)"))
# Linear Regression (full)
results.append(get_residuals_and_preds(lambda: LinearRegression(), X_full, y, "Linear (Amenities)"))
# Lasso Regression (simple)
results.append(get_residuals_and_preds(lambda: LassoCV(alphas=np.linspace(0, 5, 100), cv=5), X_simple, y, "Lasso (Simple)"))
# Lasso Regression (full)
results.append(get_residuals_and_preds(lambda: LassoCV(alphas=np.linspace(0, 5, 100), cv=5), X_full, y, "Lasso (Amenities)"))
# Ridge Regression (simple)
results.append(get_residuals_and_preds(lambda: RidgeCV(alphas=np.linspace(0.01, 5, 100)), X_simple, y, "Ridge (Simple)"))
# Ridge Regression (full)
results.append(get_residuals_and_preds(lambda: RidgeCV(alphas=np.linspace(0.01, 5, 100)), X_full, y, "Ridge (Amenities)"))

# Plotting
fig, axes = plt.subplots(len(results), 3, figsize=(18, 5 * len(results)))
fig.suptitle("Resultados: Comparación de Modelos y Residuales", fontsize=20, y=1.02)

for i, (y_test, y_pred, residuales, title) in enumerate(results):
    # Actual vs Predicted
    axes[i, 0].scatter(y_test, y_pred, alpha=0.5)
    axes[i, 0].plot(y_test, y_test, color='red', linestyle='-')
    axes[i, 0].set_title(f"{title} - Actual vs Predicted")
    axes[i, 0].set_xlabel('Actual Values')
    axes[i, 0].set_ylabel('Predicted Values')

    # Residuals Histogram
    axes[i, 1].hist(residuales, bins=30, edgecolor='black', alpha=0.7)
    mean, std_dev = np.mean(residuales), np.std(residuales)
    x = np.linspace(min(residuales), max(residuales), 100)
    y_norm = norm.pdf(x, mean, std_dev)
    y_norm = y_norm * len(residuales) * np.diff(np.histogram(residuales, bins=30)[1])[0]
    axes[i, 1].plot(x, y_norm, 'r--', label='Distribución Normal')
    axes[i, 1].axvline(x=0, color='r', linestyle='--')
    axes[i, 1].set_title(f"{title} - Histograma de Residuales")
    axes[i, 1].set_xlabel('Residuales')
    axes[i, 1].set_ylabel('Frecuencia')

    # QQ Plot
    stats.probplot(residuales, dist="norm", plot=axes[i, 2])
    axes[i, 2].set_title(f"{title} - QQ Plot")

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
